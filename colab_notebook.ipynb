```python
# Check results
import os
import pandas as pd

print('Real audio files:')
real_count = len(os.listdir('data/real')) if os.path.exists('data/real') else 0
print(f'  Total: {real_count}')

print('\nFake audio files:')
for engine in ['silero', 'coqui', 'ruslan', 'mailabs']:
    path = f'data/fake/{engine}'
    count = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f'  {engine}: {count}')

# Show metadata
if os.path.exists('metadata.csv'):
    df = pd.read_csv('metadata.csv')
    print(f'\nMetadata shape: {df.shape}')
    print('\nFirst 5 rows:')
    print(df.head())
```

```python
# Optional: Copy results to Google Drive
import shutil

drive_path = '/content/drive/MyDrive/audio_preparation_results'
os.makedirs(drive_path, exist_ok=True)

# Copy data and metadata
if os.path.exists('data'):
    shutil.copytree('data', f'{drive_path}/data', dirs_exist_ok=True)
    print(f'Copied data/ to Drive: {drive_path}/data')

if os.path.exists('metadata.csv'):
    shutil.copy('metadata.csv', f'{drive_path}/metadata.csv')
    print(f'Copied metadata.csv to Drive')
```
